```python
def pattern1():
    for i in range(1, 6):
        print((str(i) + ' ') * i)

pattern1()
```

```python
def pattern2():
    for i in range(1, 6):
        print(' ' * (i - 1) + (str(i) + ' ') * (6 - i))

pattern2()
```

```python
def pattern3():
    for i in range(1, 6):
        print(' ' * (5 - i) + ' '.join(str(j) for j in range(1, i + 1)))

pattern3()
```

```python
def pattern8():
    num = 1
    for i in range(1, 6, 2):
        print(' '.join(str(num + j) for j in range(i)))
        num += i

pattern8()
```

```python
def pattern9():
    num = 1
    for i in range(1, 5):
        row = []
        for j in range(i):
            row.append(str(num))
            num += 1
        print(' '.join(reversed(row)))

pattern9()
```

```python
def pattern10():
    for i in range(1, 6):
        line = ' '.join(str(j) for j in range(1, i + 1))
        line += ' ' + ' '.join(str(j) for j in range(i - 1, 0, -1))
        print(line)

pattern10()
```

```python
def pattern11():
    for i in range(1, 6):
        left_side = ' '.join(str(6 - j) for j in range(1, 6 - i + 1))
        right_side = ' '.join(str(j) for j in range(i, 6))
        print(left_side + ' ' + right_side.strip())

pattern11()
```
